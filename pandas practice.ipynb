Series concept in pandas

```python
#Series concept in pandas
#Series is like colomn in a table ,it holds the 1-D array    
import pandas as pd
arr=[1,2,3,4]
series=pd.Series(arr)
print(series)
print(type(pd.Series(arr)))
```

```python
#creating of labels using "index"argment to the above cell
label=pd.Series(arr,index=["x","y","z","Z"])
print(label)
```

```python
#Key/Value Objects as Series
#You can also use a key/value object, like a dictionary, when creating a Series
#here no need to import pandas because alerdy imported in the starting cell
#dictionary is key value pairs but it does not allows the duplicate key,it allowes the values
#Note: The keys of the dictionary become the labels.
dictionary={"cerebra":6000,"IBM":5000,"TCS":4000}
kvo=pd.Series(dictionary) #kvo is the varible
print(kvo)
print("")
print("*** TO select the some items in the Dictionary to use the index argment ***")
#To select only some of the items in the dictionary, use the index argument and specify only the items you want to include in the Series
dictionary1=pd.Series(dictionary,index=["cerebra","TCS"])
print(dictionary1)
```

DataFrames

```python
#DataFrames
#Data sets in Pandas are usually multi-dimensional tables, called DataFrames.
#Series is like a column, a DataFrame is the whole table.
dataset={
    "cls-1":[1,2,3,4],"cls-2":[100,200,300,400],"cls-3":[800,900,1000,1200]
}
dataframe=pd.DataFrame(dataset)
print(dataframe)
```

```python
#Locate Row
#As you can see from the result above, the DataFrame is like a table with rows and columns.
#Pandas use the "loc" attribute to return one or more specified row(s)
print(dataframe.loc[0])
```

```python
#use a list of indexes:
print(dataframe.loc[[0,1]])
```

```python
# Named Indexes
Dataframe_named_index=pd.DataFrame(dataset,index=["one","two","three","four"])
print(Dataframe_named_index)
```

```python
#Locate Named Indexes
print(Dataframe_named_index.loc["one"])
```

```python
#Load Files Into a DataFrame
# https://pandas.pydata.org/pandas-docs/stable/user_guide/io.html
#If you have a large DataFrame with many rows, Pandas will only return the first 5 rows, and the last 5 rows
#use to_string() to print the entire DataFrame.

load_file = pd.read_csv('data-1.csv')
print(load_file)  
```

```python
# for printing of total dataframe to use the to_string() method
print(load_file.to_string())
```

```python
#Analyzing DataFrames
#Viewing the Data
# head() method
#The head() method returns the headers and a specified number of rows, starting from the top
print("TOP 5 ROWS from the starting")
print(load_file.head(),"\n") #if not mentchined the specified number of rows by default it will gives the top 5 rows from the starting
print("TOP 10 ROWS")
print(load_file.head(10),"\n")
print("starting 5 rows from the bottom\n",load_file.tail(),"\n")
print("starting 10 rows from the bottom\n",load_file.tail(10))
```

Pandas - Cleaning Data

```python
#Remove Rows
#One way to deal with empty cells is to remove rows that contain empty cells.
#This is usually OK, since data sets can be very big, and removing a few rows will not have a big impact on the result.

dropna_1=pd.read_csv('data.csv')
print(dropna_1.dropna().head(2))
```

```python
# If you want to change the original DataFrame, use the inplace = True argument:
# Note: Now, the dropna(inplace = True) will NOT return a new DataFrame, but it will remove all rows containg NULL values from the original DataFrame.
dropna_2=dropna_1.dropna( inplace = True )
print(dropna_2)
```

```python
# Replace Empty Values in empty cells
# The fillna() method allows us to replace empty cells with a value
# Replace NULL values with the number 130
fillna_1=pd.read_csv('data.csv')
fillna_2=fillna_1.fillna(130,inplace=True)
print(fillna_2) 
```
